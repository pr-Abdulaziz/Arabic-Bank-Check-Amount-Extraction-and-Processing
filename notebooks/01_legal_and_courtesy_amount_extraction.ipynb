```python
from pathlib import Path
from ultralytics import YOLO

cwd = Path.cwd().resolve()
repo_root = cwd.parent if cwd.name == "notebooks" else cwd

paths = {
    "images_dir": repo_root / "CheckImages",
    "yolo_data_dir": repo_root / "data" / "yolo",
    "yolo_labels_dir": repo_root / "data" / "labels" / "yolo",
    "dataset_yaml": repo_root / "data" / "yolo" / "dataset.yaml",
    "weights_dir": repo_root / "models" / "yolo",
    "runs_dir": repo_root / "outputs" / "yolo_runs",
    "courtesy_crops_dir": repo_root / "data" / "extracted" / "courtesy",
    "legal_crops_dir": repo_root / "data" / "extracted" / "legal",
}

for d in [
    paths["yolo_data_dir"],
    paths["yolo_labels_dir"],
    paths["weights_dir"],
    paths["runs_dir"],
    paths["courtesy_crops_dir"],
    paths["legal_crops_dir"],
]:
    d.mkdir(parents=True, exist_ok=True)

for k, v in paths.items():
    print(f"{k}: {v} | exists={v.exists()}")
```
